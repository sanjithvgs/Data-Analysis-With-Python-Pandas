# **10. Data Cleaning [ Grade Column ] changing to numeric value**

```python
dat.head()
```

```python
dat['Grade'].unique()
```

```python
dat['Grade']=dat['Grade'].map({'A++ ':5,'A+ ':4,'A ':3,'A- ':2,'B+ ':1})
```

```python
dat['Grade']=dat['Grade'].astype('int')
```
